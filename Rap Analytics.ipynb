For each era we will present:
   - Artists used for analyzation
   - Number of songs analyzed per artist
   - Topic being analyzed
   - Individual scores for topic per era
   - Group score for topic per era
   - Artist with highest scores for topic per era
   - Artist with lowest scores for topic per era
   - Possible offsets in data
   - Trends and possible reasoning

Topics analyzed are:
   - Use of curse words
   - Use of racial slurs
   - Use of sexual slurs
   - Most used non-filler words

## Use of curse words

```python
swear_jar = {'fuck':0, 'shit':0, 'damn':0, 'hell':0, 'crap':0}

lyrics = open('1990-1999/AllSongs/all1990-1999.txt', 'r')

for line in lyrics:
    for word in line.split():
        if(word in swear_jar):
            swear_jar[word] +=1
        
x_values = np.arange(len(swear_jar.keys()))
num_swears = swear_jar.values()

plt.bar(x_values, num_swears, align='center', alpha=0.5)
plt.xticks(x_values, swear_jar.keys())
plt.title('CURSE WORD USAGE FROM: 1990-1999')
plt.ylabel("TIMES USED")
plt.xlabel("WORDS")
plt.show()

for i in swear_jar:
    print(i, ': ', swear_jar[i])
```
